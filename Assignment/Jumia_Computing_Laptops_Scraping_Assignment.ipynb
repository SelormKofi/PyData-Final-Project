```python
from urllib.request import Request, urlopen
```

```python
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
```

```python
webpage = urlopen(req).read()
# webpage
```

```python
from bs4 import BeautifulSoup as bs # import required libraries
```

```python
soup = bs(webpage, "html.parser")
```

```python
lap_name = soup.findAll(class_="name")
#lap_name
```

```python
lap_price = soup.findAll(class_="prc")
#lap_price
```

```python
Laptop = [i.text.strip() for i in lap_name] # get all job roles
#Laptop
```

```python
Laptop_prices = [j.text.strip() for j in lap_price] # get all job roles
#Laptop_prices
```

```python
# After scraping or collecting data, there is the need to store it. 
# For this, we use pandas

import pandas as pd
```

```python
df1 = pd.DataFrame(Laptop, columns=["Laptop Brand"]) # store the data in a dataframe
#df1
```

```python
df2 = pd.DataFrame(Laptop_prices, columns=["Laptop Prices"]) # store the data in a dataframe
#df2
```

```python
frames = [df1.astype(str), df2.astype(str)]
```

```python
result = pd.concat(frames, axis=1)
```

```python
result
```

```python
# save our data in a csv file
result.to_csv("Laptops_and_Prices_from_Jumia.csv", encoding="utf-8-sig")
```
